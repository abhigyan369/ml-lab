```python
# Install required libraries before running:
# pip install pandas pgmpy scikit-learn

import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
from sklearn.model_selection import train_test_split

# -----------------------------
# Load Heart Disease Dataset
# -----------------------------

# Dataset columns
columns = [
    'age', 'sex', 'cp', 'trestbps', 'chol',
    'fbs', 'restecg', 'thalach', 'exang',
    'oldpeak', 'slope', 'ca', 'thal', 'target'
]

# Load dataset
data = pd.read_csv('/content/drive/MyDrive/heart.csv')

# Select only important columns for simplicity
data = data[['age', 'sex', 'cp', 'chol', 'target']]

# Convert continuous values into categories
data['age'] = pd.cut(data['age'],
                     bins=[20, 40, 60, 80],
                     labels=['Young', 'Middle', 'Old'])

data['chol'] = pd.cut(data['chol'],
                      bins=[100, 200, 300, 600],
                      labels=['Low', 'Medium', 'High'])

# -----------------------------
# Create Bayesian Network
# -----------------------------

model = BayesianNetwork([
    ('age', 'target'),
    ('sex', 'target'),
    ('cp', 'target'),
    ('chol', 'target')
])

# Train the model
model.fit(data, estimator=MaximumLikelihoodEstimator)

# -----------------------------
# Perform Inference
# -----------------------------

inference = VariableElimination(model)

# Example prediction
result = inference.query(
    variables=['target'],
    evidence={
        'age': 'Old',
        'sex': 1,
        'cp': 2,
        'chol': 'High'
    }
)

# -----------------------------
# Output Result
# -----------------------------

print("\nProbability of Heart Disease:\n")
print(result)
```

```python
## Adaline
'''
Adaline is a simplest simplest neuron in deep learning, which learns from its previous
output using mean square error.
3 steps
take input
multiply by weights
learn from error
perceptron- uses hard decison immediately, learn by classfn error, less accurate
adaline - uses continous value, learn from LMS error, more accurate
Wnew = Wold + learning rate * (target - output) * input
'''
```

```python
import numpy as np
```

```python
class Adaline:
   # constructor
    def __init__(self,lr=0.1, epochs=100):
      # learning rate
      self.lr = lr
      # num of iteration
      self.epochs = epochs
      # random weights
      self.w = np.random.rand(2)
      # random bias
      self.b = np.random.rand(1)

   # Training
    def fit(self, X, y):
      for epoch in range(self.epochs):
        for i in range(len(X)):
          # predicton
          output = np.dot(X[i], self.w) + self.b
          # error
          error = y[i] - output
          # update weights
          self.w = self.w + self.lr* (error) * X[i]
          # update bias
          self.b = self.b + self.lr * error


   # Predicton
    def predict(self, X):
      output = np.dot(X, self.w) + self.b
      # step function
      if output >= 0.5:
        return 1
      else:
        return 0
```

```python
# INPUTS
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# AND GATE OUTPUT
y = np.array([0,0,0,1])
```

```python
# create model
model = Adaline()
```

```python
# train model
model.fit(X,y)
```

```python
# TEST
print(model.predict([0,0]))
print(model.predict([0,1]))
print(model.predict([1,0]))
print(model.predict([1,1]))
```

```python
# FINAL WEIGHTS
print("Weights:", model.w)
print("Bias:", model.b)
```

```python
'''
Find S Algorithm- used to find most specific hyposthesis that matches
all positive training examples

It rejects negative training examples
phi = most specific hypothesis
? - any value accepeted


'''
# TRAINING DATA
data = [
    ['Sunny', 'Warm', 'Yes'],
    ['Sunny', 'Cold', 'Yes'],
    ['Rainy', 'Warm', 'No'],
    ['Sunny', 'Warm', 'Yes']
]
# initital hypothesis
hypothesis = ['0', '0']
# find s algo
for example in data:
  # only choose positive examples
  if example[-1] == "Yes":
    for i in range(len(hypothesis)):
      # first positve example
      if hypothesis[i] == '0':
        hypothesis[i] = example[i]
      # if differ
      elif hypothesis[i] != example[i]:
        hypothesis[i] = '?'
# FINAL HYPOTHESIS
print("Final Hypothesis:")
print(hypothesis)
```

```python
data = pd.read_csv('/content/drive/MyDrive/findsdata.csv')
data.head()
```

```python

# NUMBER OF ATTRIBUTES
n = len(data.columns) - 1

# INITIAL SPECIFIC HYPOTHESIS
S = ['0'] * n

# INITIAL GENERAL HYPOTHESIS
G = ['?'] * n

# TRAINING
for i in range(len(data)):

    # ATTRIBUTES
    x = data.iloc[i, :-1].tolist()

    # TARGET
    target = data.iloc[i, -1]

    # POSITIVE EXAMPLE
    if target == 'Yes':

        for j in range(n):

            # FIRST POSITIVE EXAMPLE
            if S[j] == '0':
                S[j] = x[j]

            # GENERALIZE S
            elif S[j] != x[j]:
                S[j] = '?'

    # NEGATIVE EXAMPLE
    else:

        for j in range(n):

            # SPECIALIZE G
            if x[j] != S[j]:
                G[j] = S[j]
            else:
                G[j] = '?'

# FINAL OUTPUT
print("Specific Hypothesis (S):", S)
print("General Hypothesis (G):", G)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score

# DOCUMENTS
docs = [
    "I love cricket",
    "Cricket is amazing",
    "I hate rain",
    "Rain is terrible"
]

# LABELS
# sports = 1
# weather = 0
y = [1, 1, 0, 0]

# CONVERT TEXT TO NUMBERS
cv = CountVectorizer()
X = cv.fit_transform(docs)

# MODEL
model = MultinomialNB()

# TRAIN MODEL
model.fit(X, y)

# TEST DATA
test_docs = [
    "I love rain",
    "cricket is good"
]

# CONVERT TEST DATA
X_test = cv.transform(test_docs)

# PREDICTION
pred = model.predict(X_test)

print("Predictions:", pred)

# ACTUAL OUTPUT
y_true = [0, 1]

# ACCURACY
print("Accuracy:", accuracy_score(y_true, pred))

# PRECISION
print("Precision:", precision_score(y_true, pred))

# RECALL
print("Recall:", recall_score(y_true, pred))
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# DATASET
data = {
    'Weather': ['Sunny', 'Sunny', 'Rainy', 'Rainy'],
    'Temp': ['Hot', 'Mild', 'Hot', 'Cold'],
    'Play': ['No', 'Yes', 'Yes', 'No']
}

# CREATE DATAFRAME
df = pd.DataFrame(data)

# CONVERT TEXT TO NUMBERS
df['Weather'] = df['Weather'].map({
    'Sunny': 0,
    'Rainy': 1
})

df['Temp'] = df['Temp'].map({
    'Hot': 0,
    'Mild': 1,
    'Cold': 2
})

df['Play'] = df['Play'].map({
    'No': 0,
    'Yes': 1
})

# INPUTS
X = df[['Weather', 'Temp']]

# OUTPUT
y = df['Play']

# CREATE MODEL
model = DecisionTreeClassifier(criterion='entropy')

# TRAIN MODEL
model.fit(X, y)

# NEW SAMPLE
sample = [[0, 1]]   # Sunny, Mild

# PREDICTION
prediction = model.predict(sample)

# OUTPUT RESULT
if prediction[0] == 1:
    print("Play = Yes")
else:
    print("Play = No")
```
