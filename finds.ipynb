```python
import pandas as pd #CSV file read karne ke liye.
# read csv
data = pd.read_csv("trainingdata.csv") #CSV ko DataFrame me convert karta hai
print("Training Data:\n")
print(data)
```

```python

# Total attributes except target column
attributes = data.columns[:-1]

# Initialize hypothesis with Ø
hypothesis = ['Ø'] * len(attributes)

# Traverse each training example
for i in range(len(data)):

    # If positive example
    if data.iloc[i, -1] == "Yes":

        print(f"\nProcessing Positive Example {i+1}")

        # Compare each attribute
        for j in range(len(attributes)):

            # Current attribute value
            value = data.iloc[i, j]

            # If hypothesis is empty
            if hypothesis[j] == 'Ø':
                hypothesis[j] = value

            # If values mismatch
            elif hypothesis[j] != value:
                hypothesis[j] = '?'

        print("Updated Hypothesis:")
        print(hypothesis)

# Final hypothesis
print("\nFinal Hypothesis:")
print(hypothesis)
```

```python
import pandas as pd

data = pd.read_csv("trainingdata.csv")

hypothesis = None

for row in data.values:

    if row[-1] == "Yes":

        if hypothesis is None:
            hypothesis = row[:-1].copy()

        else:

            for i in range(len(hypothesis)):

                if hypothesis[i] != row[i]:
                    hypothesis[i] = '?'

print("Final Hypothesis:")
print(hypothesis)
```
